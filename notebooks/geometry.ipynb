```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
def surface(r):
    return 4 * math.pi * r ** 2

def volume(r):
    return (4/3) * math.pi * r ** 3

def surface_sum(r):
    return sum([surface(i) for i in range(1, r+1)])
        
```

```python
x = np.arange(1, 1001)
```

```python
plt.plot(x, volume(x))
plt.plot(x, [surface_sum(i) for i in x])
```
